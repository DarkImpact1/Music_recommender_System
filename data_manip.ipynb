## Creating all the necessary functions for textual preprocessing 

```python
def convert_ms_to_min_sec(duration_ms):
    minutes = duration_ms / 60000
    return round(minutes, 2)

def fix_artists(name):
    if isinstance(name, str):
        name = name.replace(" ", "_")
        name = name.split(";")
        name = " ".join(name)
    return name

# def fix_album_name(album):
#     if isinstance(album, str):
#         album = re.sub(r'[\(\)\-]', ' ', album)
#     return album

# def fix_track_name(text):
#     if isinstance(text, str) and len(text.strip()) > 0:
#         text = contractions.fix(text)
#     return text

def create_tag(df):
    object_columns = ['artists', 'album_name', 'track_name', 'track_genre']
    df["tag"] = df['artists'] + " " + df['track_name'] + " " + df['album_name'] + " " + df['track_genre']
    df.drop(columns=object_columns, inplace=True)
    return df

```

## Transformers to apply all the above functions for textual preprocessing

```python
# Transformer to drop columns
class DropColumnsTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns_to_drop):
        self.columns_to_drop = columns_to_drop

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        return X.drop(columns=self.columns_to_drop)

# Transformer to preprocess 'duration_ms'
class DurationTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X['duration_ms'] = X['duration_ms'].apply(convert_ms_to_min_sec)
        return X

# Transformer to process text columns
class TextProcessingTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X['artists'] = X['artists'].apply(fix_artists)
        # X['album_name'] = X['album_name'].apply(fix_album_name)
        # X['track_name'] = X['track_name'].apply(fix_track_name)
        return X

# # Transformer to create 'tag' column and drop the original text columns
# class TagColumnTransformer(BaseEstimator, TransformerMixin):
#     def fit(self, X, y=None):
#         return self

#     def transform(self, X, y=None):
#         return create_tag(X)

# Transformer to create 'tag' column and process it
class TagColumnTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X = create_tag(X)
        X['tag'] = X['tag'].apply(process_tag)
        return X
```

```python
# Define the pipeline
pipeline = Pipeline(steps=[
    ('drop_columns', DropColumnsTransformer(columns_to_drop=['Unnamed: 0', 'key', 'mode', 'time_signature', 'explicit'])),
    ('process_duration', DurationTransformer()),
    ('process_text', TextProcessingTransformer()),
    ('create_and_process_tag', TagColumnTransformer())
])
```
